# 1. What is the total Item Outlet Sales?

```python
total_item_outlet_sales = df["Item_Outlet_Sales"].sum()
print(total_item_outlet_sales)
```

# 2. What is the average Item Outlet Sales?

```python
avg_item_outlet_sales = df["Item_Outlet_Sales"].mean()
print(avg_item_outlet_sales)
```

# 3. Which Item_Type generates the highest total sales?

```python
high_sales_by_item_type = (
    df.groupby("Item_Type")["Item_Outlet_Sales"]
    .sum()
    .sort_values(ascending=False)
)

print(high_sales_by_item_type)
```

# 4. Which Item_Type generates the lowest total sales?

```python
low_sales_by_item_type = (
    df.groupby("Item_Type")["Item_Outlet_Sales"]
    .sum()
    .sort_values(ascending=True)
)
print(low_sales_by_item_type)
```

# 5. What is the sales distribution across different Item_Fat_Content?

```python
fat_content_sales = df.groupby("Item_Fat_Content")["Item_Outlet_Sales"].sum().round(2)

print(fat_content_sales)
```

# 6. Do Low Fat items sell more than Regular items?

```python
# Yes, Low Fat Sell more than Regular items
fat_content_sales = df.groupby("Item_Fat_Content")["Item_Outlet_Sales"].sum().round(2)
print(fat_content_sales)
```

# 7. What is the average sales per item?

```python
avg_item_sale = df.groupby("Item_Type")["Item_Outlet_Sales"].mean().sort_values(ascending=False)
print(avg_item_sale)
```

# 8. Which Item_Identifier contributes the highest sales?

```python
item_identifier_sales = df.groupby("Item_Identifier")["Item_Outlet_Sales"].sum().sort_values(ascending=False)
print(item_identifier_sales)
```

# 9. What is the average Item_MRP across all products?


```python
avg_MRP = df['Item_MRP'].mean()
print(avg_MRP)
```

# 10. What is the relationship between Item_MRP and Item_Outlet_Sales?

```python
correlation = df["Item_MRP"].corr(df["Item_Outlet_Sales"]).round(2)

print(correlation)
```

# 11. Do higher MRP items generate higher sales?

```python
df["MRP_Group"] = pd.qcut(df["Item_MRP"], 2, labels=["Low", "High"])
df.groupby("MRP_Group")["Item_Outlet_Sales"].mean()
```

# 12. How does Item_Visibility affect sales?

```python
visibility_sales_corr = df["Item_Visibility"].corr(df["Item_Outlet_Sales"])
visibility_sales_corr
```

# 13. Are there items with low visibility but high sales?

```python
df[(df["Item_Visibility"] < 0.05) & (df["Item_Outlet_Sales"] > df["Item_Outlet_Sales"].mean())]
```

# 14. Which items have the highest visibility but low sales?

```python
df[(df["Item_Visibility"] > 0.05) & 
(df["Item_Outlet_Sales"] < df["Item_Outlet_Sales"].mean())]
```
